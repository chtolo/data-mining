# **data visualisation



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Charger les données
url = 'https://raw.githubusercontent.com/chtolo/data-mining/main/boston_housing_clean.csv'
data = pd.read_csv(url)
```

```python
# Afficher les premières lignes du DataFrame pour s'assurer que les données sont chargées correctement
print(data.head())
```

```python
# Visualisation des données
# Exemple : Diagramme en barres montrant la relation entre la variable RM (nombre moyen de pièces par logement) et la valeur médiane des logements
plt.figure(figsize=(10, 6))
plt.bar(data['crim'], data['medv'], color='skyblue')
plt.title('Relation entre le nombre moyen de pièces et la valeur médiane des logements')
plt.xlabel('Nombre moyen de pièces')
plt.ylabel('Valeur médiane des logements')
plt.show()
```

```python


# Visualisation des données
plt.figure(figsize=(10, 6))
plt.scatter(data['crim'], data['medv'], color='skyblue', alpha=0.7)
plt.title('Relation entre le nombre de crimes et les prix de l\'immobilier à Boston')
plt.xlabel('Nombre de crimes pour 10000 habitants')
plt.ylabel('Prix médian des logements en milliers de dollars')
plt.grid(True)
plt.show()
```

```python
# Visualisation des données
plt.figure(figsize=(10, 6))
plt.scatter(data['tax'], data['medv'], color='orange', alpha=0.7)
plt.title('Relation entre la taxe foncière et les prix de l\'immobilier à Boston')
plt.xlabel('Taux de taxe foncière pour 10000$')
plt.ylabel('Prix médian des logements en milliers de dollars')
plt.grid(True)
plt.show()
```
